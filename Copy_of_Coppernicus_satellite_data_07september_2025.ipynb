```python
import pygrib
import pandas as pd
import numpy as np
import os

extracted_dir = 'extracted_grib'
grib_files = [f for f in os.listdir(extracted_dir) if f.endswith('.grib')]

data = []
if grib_files:
    grib_file_path = os.path.join(extracted_dir, grib_files[0])
    grbs = pygrib.open(grib_file_path)

    for grb in grbs:
        var_name = grb.name
        values = grb.values  # 2D array
        valid_date = grb.validDate

        # Flatten the 2D array and ensure it has 2796 values
        flattened_values = values.flatten()
        if len(flattened_values) != 2796:
            print(f"Warning: {var_name} has {len(flattened_values)} values, expected 2796")
            continue

        data.append({
            'variable': var_name,
            'values': flattened_values.tolist(),
            'valid_date': valid_date
        })

    grbs.close()

    # Create initial DataFrame
    df = pd.DataFrame(data)

    # Pivot the DataFrame to have variables as columns and valid_date as index
    df_pivoted = df.pivot(index='valid_date', columns='variable', values='values')

    # Expand the lists of values (2795 per variable) into rows
    result_data = []
    for date in df_pivoted.index:
        row = {}
        for var in df_pivoted.columns:
            values = df_pivoted.loc[date, var]  # List of 2796 values
            for i in range(2796):
                row[f'{var}_value_{i+1}'] = values[i]
        row['valid_date'] = date
        result_data.append(row)

    # Create final DataFrame
    final_df = pd.DataFrame(result_data)
    final_df.set_index('valid_date', inplace=True)

    print(final_df)
else:
    print("No GRIB files found to open.")
```

```python
import pygrib
import pandas as pd
import numpy as np
import os

extracted_dir = 'extracted_grib'
grib_files = [f for f in os.listdir(extracted_dir) if f.endswith('.grib')]

data = []
if grib_files:
    grib_file_path = os.path.join(extracted_dir, grib_files[0])
    grbs = pygrib.open(grib_file_path)

    for grb in grbs:
        var_name = grb.name
        values = grb.values  # 2D array
        valid_date = grb.validDate
        flattened_values = values.flatten()  # Flatten 2D array for DataFrame compatibility
        data.append({
            'variable': var_name,
            'valid_date': valid_date,
            'values': flattened_values.tolist()
        })
    grbs.close()

    # Check if data was collected
    if data:
        # Ensure all value arrays have the same length
        max_len = max(len(item['values']) for item in data)
        if not all(len(item['values']) == max_len for item in data):
            raise ValueError("All GRIB messages must have the same number of values for pivoting.")

        # Create a DataFrame with index as the value positions
        df_list = []
        for item in data:
            # Create a unique column name combining variable and date
            col_name = f"{item['variable']}_{item['valid_date'].strftime('%Y%m%d_%H%M%S')}"
            df_list.append(pd.Series(item['values'], name=col_name))

        # Combine all series into a single DataFrame
        df = pd.concat(df_list, axis=1)
        print(df)
    else:
        print("No GRIB messages found in the file.")
else:
    print("No GRIB files found to open.")
```
